```python
from zumi.zumi import Zumi
zumi = Zumi() 
zumi.reset_gyro()
zumi.mpu.calibrate_MPU()
POWER = 40 
TIME  = 1

#use this program in order to find the 
#max and min values of your 
#front left and right IR sensors

left_list = []
right_list = []

for i in range(0,200):
    # note can be a number between 0-60
    ir_readings = zumi.get_all_IR_data()

    left_list.append(ir_readings[5])
    right_list.append(ir_readings[0])
    #print("right= " ,ir_readings[0], " left= ", ir_readings[5])

    
max_left  = max(left_list)
max_right = max(right_list)
min_left  = min(left_list)
min_right = min(right_list)
    
print("left: max ", max_left, " min ",min_left)
print("right: max ", max_right, " min ",min_right)
```

```python
'''
The objective of this program is upon detecting a hand or more generally an object at 
a medium distance, Zumi will reverse for a second and begin driving straight again.
This is the warm up...
'''
def main():
    print("Program Started")
    
    # get_all_IR_data() returns a list of 6 elements
    # this list contains the readings
    # of the 6 IR sensors
    # ir_readings[0] = front right
    # ir_readings[1] = bottom right
    # ir_readings[2] = back right
    # ir_readings[3] = bottom left
    # ir_readings[4] = back left
    # ir_readings[5] = front left
  
  
    threshold_IR_FL  = (max_left + min_left)/2
    threshold_IR_FR  = (max_right+ min_right)/2
    
    print("threshold_IR_FL",threshold_IR_FL)
    print("threshold_IR_FR",threshold_IR_FR)
    
    while True : 
        
        ir_reading = zumi.get_all_IR_data()
        front_right_ir = ir_reading[0]
        front_left_ir  = ir_reading[5]
        
        if front_left_ir < threshold_IR_FL and front_right_ir < threshold_IR_FR:
            zumi.reverse(POWER, TIME )
   
    zumi.stop()
    print("Program Ended")
try:
    main()
finally:
    zumi.stop()
```

```python
'''
The objective of this program is upon detecting a hand or more generally an object at 
a medium distance, Zumi will reverse ONLY FOR THE DURATION OF TIME THE OBJECT IS DETECTED.
'''
def getFrontSensorValues():
    ir_reading = zumi.get_all_IR_data()
    front_right_ir = ir_reading[0]
    front_left_ir  = ir_reading[5]
    return front_left_ir, front_right_ir

def main():
    print("Program Started")
    
    # get_all_IR_data() returns a list of 6 elements
    # this list contains the readings
    # of the 6 IR sensors
    # ir_readings[0] = front right
    # ir_readings[1] = bottom right
    # ir_readings[2] = back right
    # ir_readings[3] = bottom left
    # ir_readings[4] = back left
    # ir_readings[5] = front left
  
  
    threshold_IR_FL  = (max_left + min_left)/2
    threshold_IR_FR  = (max_right+ min_right)/2
    
    print("threshold_IR_FL",threshold_IR_FL)
    print("threshold_IR_FR",threshold_IR_FR)
    
    while True : 
        front_left_ir, front_right_ir     = getFrontSensorValues()
        while front_left_ir < threshold_IR_FL and front_right_ir < threshold_IR_FR:
            zumi.go_reverse(40,0)
            front_left_ir, front_right_ir = getFrontSensorValues()
        zumi.go_straight(40,0)
        
    zumi.stop()
    print("Program Ended")
try:
    main()
finally:
    zumi.stop()
```
